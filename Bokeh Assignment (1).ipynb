# Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure,output_notebook , show


output_notebook()

plot = figure(title = "Scatter Plot")

plot.circle([1,2,3,4] , [5,3,6,2] ,size = 10 , color = "red" )

show(plot)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
from bokeh.plotting import figure,output_notebook, show


output_notebook()

plot = figure(title = "glyphs")

plot.circle([1,2,3,4] , [5,3,5,6] , size = 10 , color = "blue")
plot.line([1,2,3,4] , [1,4,3,5] , line_width = 10 , color = "red")

show(plot)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure, show , output_notebook

output_notebook()


plot = figure(width=400, height=300, title="Customized Plot")

plot.xaxis.axis_label = "X-axis"
plot.yaxis.axis_label = "Y-axis"
plot.xaxis.major_label_text_font = "Arial"
plot.yaxis.major_label_text_font = "Arial"
plot.x_range.range_padding = 0.1
plot.y_range.range_padding = 0.1

plot.title.text = "My Custom Plot"
plot.title.text_font = "Verdana"
plot.title.text_font_size = "18pt"

plot.circle([1, 2, 3], [4, 5, 6], size=10, color="blue", legend_label="Data Points")
plot.legend.title = "Legend"
plot.legend.label_text_font = "Helvetica"
plot.legend.label_text_font_size = "12pt"

show(plot)
```

# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a Python server that allows you to create and deploy interactive Bokeh plots and applications. It provides a way to build web applications that can dynamically update and respond to user interactions or data changes in real time. The Bokeh server enables you to create interactive plots that go beyond static visualizations by incorporating user input, streaming data, and reactive behavior.

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.io import output_notebook, show

output_notebook()


plot = figure(width=400, height=300, title="Real-time Plot")

source = ColumnDataSource(data=dict(x=[], y=[]))
plot.circle('x', 'y', source=source, size=10)

def update_data():
    new_data = dict(x=[1, 2, 3, 4], y=[5, 3, 6, 2])
    source.data = new_data

update_data()
layout = row(plot)
curdoc().add_root(layout)
show(layout)
```
